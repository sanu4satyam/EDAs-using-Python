# Technique 1, Dropping all Nulls

```python
#Dropping Null values and Saving in a new Dataframe

df_drop = df.dropna()
```

```python
#Shape of the dataframe after dropping the Nulls

df_drop.shape
```

```python
#Chart after dropping all null values

plt.figure(figsize = (15,12))
sns.heatmap(df_drop.isnull())
```

```python
#Checking again Null Values

df_drop.isnull().sum()
```

```python
# Finding all the Numerical Columns

df.select_dtypes(include=['float64', 'int64']).columns
```

```python
#Making a list of the Numerical Columns

list1 = ['total_bill', 'tip', 'size']
```

```python
# Plotting the distplot for comparing Original and Dropped

plt.figure(figsize=(10,8))
for i, var in enumerate(list1):
  plt.subplot(3,1,i+1)
  sns.distplot(df[var], bins = 10)
  sns.distplot(df_drop[var], bins = 10)
  plt.legend(['Original','Dropped'])
```

In the Technique 1, I have dropped all the rows containing NULL values.
The Original Dataset has shape (244,7), And after dropping it has (222, 7).

# Technique 2, Filling with Mean

```python
df.head()
```

```python
# Finding all the Numerical Columns


df.select_dtypes(include = ['int64','float64']).columns
```

```python
#Making a list of the Numerical Columns

num_cols = ['total_bill', 'tip', 'size']
```

```python
#Filling the missing values with mean of the respective columns

df_mean = df.fillna(df.mean())
df_mean.head()
```

```python
#Checking any Null values filled or not


df_mean.isnull().sum()
```

```python
# Plotting the distplot for comparing Original and Mean

plt.figure(figsize=(10,8))
for i, var in enumerate(num_cols):
  plt.subplot(3,1,i+1)
  sns.distplot(df[var], bins = 10, kde_kws = {'color' :'black', 'linewidth':3})
  sns.distplot(df_mean[var], bins = 10, kde_kws = {'color' :'red', 'linewidth':2})
  plt.legend(['Original','Mean'])
```

# Technique 3, Filling with Mode

```python
#Getting the info the Dataframe

df.info()
```

```python
# Finding all the Numerical Columns

df.select_dtypes(include = ['object']).columns
```

```python
#Creating a list of Categorical columns.

cat_cols = ['sex', 'smoker', 'day', 'time']
```

```python
#Copying the Original Dataframe and saving it in new dataframe to perform Mode

df_mode = df.copy()
```

```python
#Checking the Head of the copied dataframe

df_mode.head()
```

```python
#Filing the Mode values in categorical columns

for var in cat_cols:
  df_mode[var].fillna(df_mode[var].mode()[0], inplace = True)
  #print(var, "=", data3_copy[var].mode()[0])
```

```python
#Checking the row is filled with mode or not

df_mode.isna().sum()
```

```python
#Plotting Chart to view the distribution between Mode and Original Data(Dropped Nulls)

plt.figure(figsize = (10,10))
for i,var in enumerate(cat_cols):
  plt.subplot(2,2, i+1)
  plt.hist(df_mode[var], label = "Mode",color = 'Cyan')
  plt.hist(df[var].dropna(), label = 'Original', color = 'Black')
  plt.legend()
```
