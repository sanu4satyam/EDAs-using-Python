```python
df_country = pd.read_excel('/content/drive/MyDrive/Zomato Datasets/Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
pd.merge(df, df_country, on= 'Country Code', how='left')
```

```python
final_df = pd.merge(df, df_country, on= 'Country Code', how='left')
```

```python
final_df.head(2)
```

```python
#To Check Datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names =final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
## Pie Chart - Top 3 Countries that uses Zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%');
```

**Observation** : Zomato's maximum transactions are from India following USA & UK in Top 3 countries

```python
ratings =final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

**Observations:**

*   When the Ratings are between 4.5 to 4.9 ----> Excellent
*   When the Ratings are between 4.0 to 4.4 ----> Very Good
*   When the Ratings are between 3.5 to 3.9 ----> Good
*   When the Ratings are between 2.5 to 3.4 ----> Average
*   When the Ratings are between 1.8 to 2.4 ----> Poor






 





```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (14,7)
sns.barplot(x="Aggregate rating",y="Rating Count", data=ratings);
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','olive']);
```

**Observations:**

*   Not Rated Count is very High
*   Maximum numbers of rating are between 2.5 to 4.1

```python
## Count Plot

sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','olive']);
```

```python
## Countries that has given 0 rating

final_df.columns
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

**Observation:** Maximum numbers of 0 ratings are from Indian Customers

```python
## Currencies used by Each Countries for Zomato Deliveries

final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
## Countries having Online Deliveries options

final_df[final_df['Has Online delivery'] == 'Yes'].Country.value_counts().reset_index()
```

**Observation:** Online Deliveries are available in India and UAE only

```python
## Top 5 Cities
city_labels =final_df.City.value_counts().index
city_values =final_df.City.value_counts().values
plt.pie(city_values[:5], labels=city_labels[:5], autopct='%1.2f%%');
```
